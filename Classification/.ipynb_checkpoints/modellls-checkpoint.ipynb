```python



##############################################################33
##############################################################
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
import torchvision.models as models
from torchvision.datasets import ImageFolder
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt

# Define the device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load the saved model state dict
path = 'preTrainedVit_50_epochs_.pth'
pretrained_vit_state_dict = torch.load(path, map_location=device)

# Setup a ViT model instance with pretrained weights
pretrained_vit = models.vit_b_16().to(device)

# Rename the keys in the loaded state dict to match the keys in the model's state dict
new_pretrained_vit_state_dict = {}
for k, v in pretrained_vit_state_dict.items():
    if 'head' in k:
        k = k.replace('head', 'heads')  # Rename keys related to the classifier head
    new_pretrained_vit_state_dict[k] = v

# Load the modified state dict into the model
pretrained_vit.load_state_dict(new_pretrained_vit_state_dict, strict=False)

# Freeze the base parameters
for parameter in pretrained_vit.parameters():
    parameter.requires_grad = False

# Change the classifier head
class_names = ['benign', 'malignant', 'normal']
pretrained_vit.conv_proj.in_channels = 1
pretrained_vit.heads = nn.Linear(in_features=768, out_features=len(class_names)).to(device)

# Define data transforms for your validation dataset
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Adjust size as per the input size of your model
    transforms.ToTensor(),
])

# Load your validation dataset
val_dataset = ImageFolder(r'C:\Users\bhara\OneDrive\Desktop\Temp\AUGMENTED DATASET\test', transform=transform)

# Create a DataLoader for the validation dataset
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

# Define lists to store ground truth labels and predictions
all_labels = []
all_preds = []

# Make predictions on the validation dataset
for images, labels in val_loader:
    with torch.no_grad():
        images = images.to(device)
        outputs = pretrained_vit(images)
        _, predicted = torch.max(outputs, 1)
        all_labels.extend(labels.tolist())  # Store true labels
        all_preds.extend(predicted.tolist())

# Calculate the confusion matrix
conf_matrix = confusion_matrix(all_labels, all_preds)

# Calculate classification report
target_names = ['benign', 'malignant', 'normal']
print(classification_report(all_labels, all_preds, target_names=target_names))

# Visualize the confusion matrix with numbers and colors
plt.figure(figsize=(8, 6))
plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)

# Add numbers in the cells
for i in range(len(class_names)):
    for j in range(len(class_names)):
        plt.text(j, i, f'{conf_matrix[i, j]}', horizontalalignment='center', verticalalignment='center', color='white')

# Add colorbar
plt.colorbar()

plt.title('Confusion Matrix')
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=45)
plt.yticks(tick_marks, class_names)
plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
import cv2
import numpy as np
import os
import random

# Define the main directory containing your original images
main_dir = '../Working/OverlayedImages1'

# Define the directory where augmented images will be saved
output_dir = '../Working'

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Dictionary to store the counter for each class label
class_counters = {}

# Function to perform rotation
def rotate_image(image, angle):
    height, width = image.shape[:2]
    rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height), borderMode=cv2.BORDER_REFLECT)
    return rotated_image
#     return image.rotate(angle, expand=True)

# Function to perform horizontal flip
def flip_image_horizontal(image):
    return cv2.flip(image, 1)

# Function to perform vertical flip
def flip_image_vertical(image):
    return cv2.flip(image, 0)

# Function to save augmented images in the desired directory structure with specified file names
def save_augmented_image(output_dir, class_label, filename, augmented_image):
    # Initialize counter for the class label if it doesn't exist
    if class_label not in class_counters:
        class_counters[class_label] = 1
    # Create directory structure in the output directory
    output_subdir = os.path.join(output_dir, 'augmented', class_label)
    os.makedirs(output_subdir, exist_ok=True)
    # Save augmented image with the specified file name
    save_filename = f'{class_label}{class_counters[class_label]}.png'
    cv2.imwrite(os.path.join(output_subdir, save_filename), augmented_image)
    # Increment counter for the class label
    class_counters[class_label] += 1

# Iterate through original images
for root, dirs, files in os.walk(main_dir):
    for filename in files:
        if filename.endswith(('.jpg', '.jpeg', '.png')):  # Check if file is an image
            # Load original image
            image = cv2.imread(os.path.join(root, filename))
            
            # Check if image is loaded successfully
            if image is None:
                print(f"Failed to load {filename}")
                continue
            
            # Extract class label from the original directory name
            class_label = os.path.basename(root)
            
            # Apply augmentations
            for i in range(2):  # Generate 10 augmented images per original image
                augmented_image = image.copy()
                # Apply horizontal flip only for the first augmented image
                if i == 0:
                    augmented_image = flip_image_horizontal(augmented_image)
                elif i == 1:
                    augmented_image = flip_image_vertical(augmented_image)
                else:
                    # Randomly select augmentation
                    augmented_image = image.copy()
                save_augmented_image(output_dir, class_label, f'{filename}', augmented_image)
```
